### Data split

```python
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=11)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=11)

df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

y_train = df_train.loan_status.values
y_val = df_val.loan_status.values
y_test = df_test.loan_status.values

del df_train['loan_status']
del df_val['loan_status']
del df_test['loan_status']
```

```python
df_train.shape , df_val.shape , df_test.shape
```

```python
dv = DictVectorizer(sparse=True)
X_train = dv.fit_transform(df_train.to_dict(orient='records'))
X_val = dv.transform(df_val.to_dict(orient='records'))
X_test = dv.transform(df_test.to_dict(orient='records'))
```

```python
y_train.shape, y_val.shape, y_test.shape
```

## Logistic Resgression

```python
lg = LogisticRegression()
lg.fit(X_train, y_train)
```

```python
y_pred = lg.predict(X_val)
```

```python
acc = accuracy_score(y_val, y_pred)
prec = precision_score(y_val, y_pred)
rec = recall_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred)
y_pred_proba = lg.predict_proba(X_val)[:, 1]
fpr, tpr, thresholds = roc_curve(y_val, y_pred_proba)
roc_auc = roc_auc_score(y_val, y_pred)
```

```python
roc_auc
```

```python
lg = LogisticRegression(solver='liblinear', penalty='l1')
lg.fit(X_train, y_train)
```

```python
def evaluate_model(model, X_val, y_val):
    y_pred = model.predict(X_val)
    acc = accuracy_score(y_val, y_pred)
    prec = precision_score(y_val, y_pred)
    rec = recall_score(y_val, y_pred)
    f1 = f1_score(y_val, y_pred)
    y_pred_proba = model.predict_proba(X_val)[:, 1]
    roc_auc = roc_auc_score(y_val, y_pred_proba)

    scores = {
    'accuracy': acc,
    'precision': prec,
    'recall_score': rec,
    'f1_score': f1,
    'roc_auc_score': roc_auc
    }

    return scores
```

```python
scores = evaluate_model(lg, X_val, y_val)
```

```python
scores
```

```python
# no of itterations alone makes no difference
lg = LogisticRegression(solver='liblinear', penalty='l1', max_iter=200)
lg.fit(X_train, y_train)
scores = evaluate_model(lg, X_val, y_val)
scores
```

```python
# no of itterations alone made a difference when class_weight parameter was added
lg = LogisticRegression(solver='liblinear', penalty='l1', class_weight='balanced', max_iter=200)
lg.fit(X_train, y_train)
scores = evaluate_model(lg, X_val, y_val)
scores
```

```python
C_values = [0.01, 0.1, 1, 10]
scores_c = {}

for C in C_values:
    model = LogisticRegression(
        solver='liblinear',
        penalty='l1',
        class_weight='balanced',
        max_iter=2000,
        C=C
    )
    model.fit(X_train, y_train)
    scores = evaluate_model(model, X_val, y_val)
    scores_c[C] = scores  
```

```python
columns = ['C_value', 'accuracy', 'precision', 'recall_score', 'f1_score', 'roc_auc_score']
df_scores = pd.DataFrame(scores_c, columns = columns)
```

```python
df_scores = pd.DataFrame.from_dict(scores_c, orient='index')
df_scores.index.name = 'C'

df_scores
```

> best final LogisticRegression Model

```python
lg = LogisticRegression(solver='liblinear', penalty='l1', class_weight='balanced', max_iter=200, C=1)
lg.fit(X_train, y_train)
scores = evaluate_model(lg, X_val, y_val)
scores
```

## Decision Tree

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict_proba(X_train)[:, 1]
auc = roc_auc_score(y_train, y_pred)
print('train', auc)
 
y_pred = dt.predict_proba(X_val)[:, 1]
auc = roc_auc_score(y_val, y_pred)
print('val', auc)

y_pred = dt.predict_proba(X_test)[:, 1]
auc = roc_auc_score(y_test, y_pred)
print('test', auc)
```

```python
dt = DecisionTreeClassifier(max_depth=1)
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict_proba(X_train)[:, 1]
auc = roc_auc_score(y_train, y_pred)
print('train', auc)
 
y_pred = dt.predict_proba(X_val)[:, 1]
auc = roc_auc_score(y_val, y_pred)
print('val', auc)
```

```python
scores = []
 
for d in [4, 5, 6]:
    for s in [1, 2, 5, 10, 15, 20, 100, 200, 500]:
        dt = DecisionTreeClassifier(max_depth=d, min_samples_leaf=s)
        dt.fit(X_train, y_train)
 
        y_pred = dt.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_val, y_pred)
         
        scores.append((d, s, auc))
 
columns = ['max_depth', 'min_samples_leaf', 'auc']
df_scores = pd.DataFrame(scores, columns=columns)
df_scores
```
